<h4>Import of libraries

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline  
```

<h4>Load transactions

```python
myfile = 'transactions.csv'
transactions = pd.read_csv(myfile, sep=';', header=0)
transactions.loc[:,'blockTimestamp'] = pd.to_datetime(transactions['blockTimestamp'])
```

<h4>Aggregate of data for addresses "from"

```python
address_from_blockTimestamp = transactions.groupby('from')['blockTimestamp'].agg({'count':np.count_nonzero, 
                                                                                  'delta_timestamp': lambda x: 
                                                                                       (np.max(x) - np.min(x)).total_seconds(),
                                                                                 'min_timestamp':np.min,
                                                                                 'max_timestamp':np.max}).reset_index()

address_from_value = transactions.groupby('from')['value'].agg({'count_value_non_zero':np.count_nonzero, 
                                                                'mean_value':np.mean, 
                                                                'max_value':np.max, 
                                                                'sum_value':np.sum}).reset_index()
```

<h4>Aggregate of data for addresses "to"

```python
address_to_blockTimestamp = transactions.groupby('to')['blockTimestamp'].agg({'count':np.count_nonzero, 
                                                                              'delta_timestamp': lambda x: 
                                                                                       (np.max(x) - np.min(x)).total_seconds(),
                                                                              'min_timestamp':np.min,
                                                                              'max_timestamp':np.max}).reset_index()

address_to_value = transactions.groupby('to')['value'].agg({'count_value_non_zero':np.count_nonzero, 
                                                            'mean_value':np.mean, 
                                                            'max_value':np.max, 
                                                            'sum_value':np.sum}).reset_index()

address_is_contract = transactions.groupby('to')['input'].agg({'is_contract': lambda x: 
                                                               False if len(np.unique(x)) == 1 
                                                                        and np.unique(x) == ['0x'] else True}).reset_index()
```

<h4>Merging aggregated DataFrames to one.  <br>

```python
address_from = pd.merge(address_from_blockTimestamp, 
                        address_from_value,
                        on='from')

address_to = pd.merge(pd.merge(address_to_blockTimestamp, 
                               address_to_value,
                               on='to'),
                      address_is_contract,
                      on='to')

addresses_analysis = pd.merge(address_from, 
                              address_to,
                              how='outer', 
                              left_on='from', 
                              right_on='to', 
                              suffixes=('_from', '_to'))
```

<h4>Calculating of address field

```python
def address_function(x):
    return x['from'] if x['from']==x['from'] else x['to']

addresses_analysis.loc[:,'address'] = addresses_analysis.apply(address_function, axis=1)
addresses_analysis.drop(['from','to'], inplace=True, axis=1)
addresses_analysis.fillna(0, inplace=True)
addresses_analysis.loc[:,'count_value_zero_from'] = addresses_analysis['count_from'] - addresses_analysis['count_value_non_zero_from']
addresses_analysis.loc[:,'count_value_zero_to'] = addresses_analysis['count_to'] - addresses_analysis['count_value_non_zero_to']
addresses_analysis.loc[:, 'percent_count_value_zero_from'] = addresses_analysis['count_value_zero_from']/addresses_analysis['count_from']
addresses_analysis.loc[:, 'percent_count_value_zero_to'] = addresses_analysis['count_value_zero_to']/addresses_analysis['count_to']
addresses_analysis.loc[:,'delta_value'] = addresses_analysis[ 'sum_value_to'] - addresses_analysis[ 'sum_value_from']
addresses_analysis.loc[:,'max_timestamp_to'] = pd.to_datetime(addresses_analysis['max_timestamp_to'])
addresses_analysis.loc[:,'max_timestamp_from'] = pd.to_datetime(addresses_analysis['max_timestamp_from'])
addresses_analysis.loc[:,'min_timestamp_to'] = pd.to_datetime(addresses_analysis['min_timestamp_to'])
addresses_analysis.loc[:,'min_timestamp_from'] = pd.to_datetime(addresses_analysis['min_timestamp_from'])

addresses_analysis.loc[:,'class_templary'] = (addresses_analysis[['max_timestamp_from', 
                                                                  'min_timestamp_from',
                                                                  'max_timestamp_to', 
                                                                  'min_timestamp_to']].max(axis=1) - 
                                              addresses_analysis[['max_timestamp_from', 
                                                                  'min_timestamp_from',
                                                                  'max_timestamp_to', 
                                                                  'min_timestamp_to']].min(axis=1)).map(lambda x: 1 if x.total_seconds()<3600 else 0)

addresses_analysis.loc[:,'class_templary_48h'] = (addresses_analysis[['max_timestamp_from', 
                                                                      'min_timestamp_from',
                                                                      'max_timestamp_to', 
                                                                      'min_timestamp_to']].max(axis=1) - 
                                                  addresses_analysis[['max_timestamp_from', 
                                                                      'min_timestamp_from',
                                                                      'max_timestamp_to', 
                                                                      'min_timestamp_to']].min(axis=1)).map(lambda x: 1 if x.total_seconds()<=3600*48 else 0)
```

<h2> First iteration. Core.

```python
classes = [0, 1, 2, 3, 4, 5]
classes_values = [500, 1000, 2000, 3000, 5000, 10000]
addresses_analysis_classes = []
addresses_analysis_classes_series = []

for i in classes:
    addresses_analysis_classes.append(addresses_analysis[
                                        ((addresses_analysis['sum_value_to']>=classes_values[i]-10) & 
                                         (addresses_analysis['sum_value_to']<=classes_values[i]+10)) &
                                        ((addresses_analysis['sum_value_from']>=classes_values[i]-10) & 
                                         (addresses_analysis['sum_value_from']<=classes_values[i]+10))]['address'])

    addresses_analysis_classes_series.append(pd.Series(index = list(addresses_analysis_classes[-1])))
    print('Count of addresses by class {:>6}: {:>10,}'.format(classes_values[i],len(addresses_analysis_classes[-1])))
```

```python
def class_calculation_1(x):
    for i in classes: 
        if x in addresses_analysis_classes_series[i]: return i
    return None
```

```python
transactions.loc[:,'class_to'] = transactions['to'].map(lambda x: class_calculation_1(x))
transactions.loc[:,'class_from'] = transactions['from'].map(lambda x: class_calculation_1(x))
addresses_analysis.loc[:,'class'] = addresses_analysis['address'].map(lambda x: class_calculation_1(x))
```

```python
def calculate_instrasting_transactions(classes):

    intresting_transactions = transactions[(transactions['class_to'].isin(classes)) | 
                                           (transactions['class_from'].isin(classes))]

    intresting_transactions_rotate = intresting_transactions[(intresting_transactions['class_to'].isin(classes)) & 
                                                             (intresting_transactions['class_from'].isin(classes))]

    intresting_transactions_input = intresting_transactions[(intresting_transactions['class_to'].isin(classes)) & 
                                                            (~intresting_transactions['class_from'].isin(classes))]

    intresting_transactions_output = intresting_transactions[(~intresting_transactions['class_to'].isin(classes)) & 
                                                             (intresting_transactions['class_from'].isin(classes))]
    print ('The Mixer Rotate, count of transactions: {:>12,}'.format(len(intresting_transactions_rotate)))
    print ('The Mixer Input, count of transactions:  {:>12,}'.format(len(intresting_transactions_input)))
    print ('The Mixer Output, count of transactions: {:>12,}'.format(len(intresting_transactions_output)))
    print ('All, count of transactions:              {:>12,}'.format(len(transactions)))

    print ('The Mixer Share in count of transactions: {:.2%}'.format((len(intresting_transactions_rotate)+
                                                            len(intresting_transactions_input)+
                                                            len(intresting_transactions_output))/
                                                        len(transactions)),'\n')

    print ('The Mixer Rotate, sum value of transactions: {:>16,.2f} ETH'.format(sum(intresting_transactions_rotate['value'])))
    print ('The Mixer Input, sum value of transactions:  {:>16,.2f} ETH'.format(sum(intresting_transactions_input['value'])))
    print ('The Mixer Output, sum value of transactions: {:>16,.2f} ETH'.format(sum(intresting_transactions_output['value'])))
    print ('All, sum value of transactions:              {:>16,.2f} ETH'.format(sum(transactions['value']), 2))

    print ('The Mixer Share in sum value of transactions: {:.2%}'.format((sum(intresting_transactions_rotate['value'])+
                                                                  sum(intresting_transactions_input['value'])+
                                                                  sum(intresting_transactions_output['value']))/
                                                               sum(transactions['value'])))
    return [intresting_transactions, 
           intresting_transactions_rotate, 
           intresting_transactions_input, 
           intresting_transactions_output]
```

```python
classes = [0, 1, 2, 3, 4, 5]
intresting_transactions, intresting_transactions_rotate, intresting_transactions_input, intresting_transactions_output = calculate_instrasting_transactions(classes)
```

<h2> Second iteration. Shell.

```python
input_addresses = intresting_transactions_input['from'].unique()
output_addresses = intresting_transactions_output['to'].unique()
addresses_analysis_input_Series = pd.Series(index = list(input_addresses))
addresses_analysis_output_Series = pd.Series(index = list(output_addresses))

addresses_analysis_input_mixer = addresses_analysis[(addresses_analysis['address'].isin(input_addresses)) 
                                                  & (addresses_analysis['class_templary']==1)]['address']
addresses_analysis_output_mixer = addresses_analysis[(addresses_analysis['address'].isin(output_addresses)) 
                                                   & (addresses_analysis['class_templary']==1)]['address']

addresses_analysis_input_mixer_Series = pd.Series(index = list(addresses_analysis_input_mixer))

addresses_analysis_output_mixer_Series =  pd.Series(index = list(addresses_analysis_output_mixer))
```

```python
def class_calculation_2(x):
    if x in addresses_analysis_input_mixer_Series:  return 7
    if x in addresses_analysis_output_mixer_Series: return 8
    return None
```

```python
transactions.loc[transactions[transactions['class_to']!=transactions['class_to']].index,
                 'class_to'] = transactions['to'].map(lambda x: class_calculation_2(x)) 
transactions.loc[transactions[transactions['class_from']!=transactions['class_from']].index,
                 'class_from'] = transactions['from'].map(lambda x: class_calculation_2(x)) 
addresses_analysis.loc[addresses_analysis[addresses_analysis['class']!=addresses_analysis['class']].index,
                       'class'] = addresses_analysis['address'].map(lambda x: class_calculation_2(x))
```

```python
classes = [0,1,2,3,4,5,7,8]
intresting_transactions_2, intresting_transactions_rotate_2, intresting_transactions_input_2, intresting_transactions_output_2 = calculate_instrasting_transactions(classes)
```

```python
input_addresses_2 = intresting_transactions_input_2['from'].unique()
output_addresses_2 = intresting_transactions_output_2['to'].unique()

input_mixer_addresses_2 = addresses_analysis[(addresses_analysis['address'].isin(input_addresses_2)) 
                                           & (addresses_analysis['class_templary']==1)]['address']

output_mixer_addresses_2 = addresses_analysis[(addresses_analysis['address'].isin(output_addresses_2)) 
                                            & (addresses_analysis['class_templary']==1)]['address']

input_mixer_addresses_2_Series = pd.Series(index = list(input_mixer_addresses_2))
output_mixer_addresses_2_Series = pd.Series(index = list(output_mixer_addresses_2))
```
